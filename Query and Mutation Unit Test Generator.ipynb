# GraphQL QUERY UNIT TEST GENERATOR #

```python
# export const MUTUAL_FRIENDS = gql`
# query getMutualFriends($user_id1 : ID!, $user_id2 : ID!){
#   getMutualFriends(user_id1: $user_id1 , user_id2 : $user_id2){
#             status
#           }
#   }
# `


api_name = "MUTUAL_FRIENDS"                             #api import name
api_call_name = "getMutualFriends"                         #api call name


test_component_name = api_call_name+"CCQT"                     #component name

input_list = """
user_id1
user_id2
"""


output_list = """
status
    """

api_parameters = input_list.split()       #api input param
api_parameters_value = [elem + "_input" for elem in input_list.split()]    #api immput param value

api_returns = output_list.split()      #return values parameter by api
returns_list_value = [elem+"_output" for elem in output_list.split()] #return values by api
```

```python
def create_parameter_list(ap,apv,comma):
    delem =   " , " if comma==True else " "
    sep = " : " if comma==True else " = "
    st = ""
    for idx in range(len(ap)):
        st += " "*20 + ap[idx] + sep + "\""+apv[idx] + "\""+ delem + "\n"
    return st[:-2]
    
def create_error_param_list(ap,apv,comma):
    er = []
    for elem in apv:
        er.append("error_"+elem)
    return create_parameter_list(ap,er,comma)

def create_component_ref(test_component_name,param_list):
    return "<" + test_component_name+" \n"+ param_list +"\n           />\n"

def create_responce_text_list(rlv):
    st=""
    for elem in rlv:
        st += "\""+ elem + "\" ,\n"+" "*10
    return st

print("""describe('"""+api_name+""" Unit Test', () => {

    it("renders without error", () => {
        renderer.create(
          <MockedProvider mocks={mocks} addTypename={false}>
          """
      +
      create_component_ref(test_component_name,create_parameter_list(api_parameters,api_parameters_value,False))
      +
            
          """</MockedProvider>
        );
      });
      
      it("should render loading state initially", () => {
        const component = renderer.create(
          <MockedProvider mocks={[]}>
            """
      +
      create_component_ref(test_component_name,create_parameter_list(api_parameters,api_parameters_value,False))
      +
      """
          </MockedProvider>
        );
        const tree = component.toJSON();
        expect(tree.children).toContain("Loading...");
      });
      
      it("should render api responce", async () => {
        const component = renderer.create(
          <MockedProvider mocks={mocks} addTypename={false}>
      """
      +
      create_component_ref(test_component_name,create_parameter_list(api_parameters,api_parameters_value,False))
      +
      """
          </MockedProvider>
        );
      
        await wait(0); // wait for response
      
        const p = component.root.findByType("Text");
       
        expect(p.children).toMatchObject([
        
          """
      +
      create_responce_text_list(returns_list_value)
      +
      """
        ]);
      });
      it("should show error UI", async () => {
        const component = renderer.create(
          <MockedProvider mocks={mocks} addTypename={false}>
            """
      +
      create_component_ref(test_component_name,create_error_param_list(api_parameters,api_parameters_value,False))
      +
      """
          </MockedProvider>
        );
      
        await wait(0); // wait for response
      
        const tree = component.toJSON();
        expect(tree.children).toContain("Error!");
      });


  })""")
```

```python
def prop_list(pl):
    st = ""
    for elem in pl:
        st+= " "+elem+" ,"
    return st
    
def prop_type(pl):
    st = ""
    for elem in pl:
        st+= " "+elem+": string ,"
    return st
def create_responce_text(rlv,ar):
    st=""
    for elem in rlv:
        st += "{data."+ar+"."+elem + "}\n"+" "*10
    return st[:-11]

print(
    """export const """+test_component_name+""" = ({ """+prop_list(api_parameters)+""" }: { """+prop_type(api_parameters)+"""  }) => (
  <Query query={"""+api_name+"""} variables={{ """+prop_list(api_parameters)+""" }}>
    {({ loading, error, data }) => {
      if (loading) return <Text>Loading...</Text>
      if (error) return <Text>Error!</Text>
      return (
        <Text>
          """
    +
            create_responce_text(api_returns,api_call_name)
    +
          """
        </Text>
      )
    }}
  </Query>
)""")
```

```python


print("""{
        request:{
            query: """+api_name+""",
            variables: {
"""
      +
                  create_parameter_list(api_parameters,api_parameters_value,True)
      +
      """
            }
        },
        result :{
            data: {
                """+api_call_name+""": {
"""
      +
                  create_parameter_list(api_returns,returns_list_value,True)
      +
      """
                }
              }
        }
    },
    {
        request:{
            query: """+api_name+""",
            variables: {
"""
      +
                  create_error_param_list(api_parameters,api_parameters_value,True)
      +
      """
            }
        },
        error: new Error("Wrong ID!")
    },""")
```

```python
# export const UPDATE_LAST_SEEN_RECIEVER = gql`
# mutation updateLastSeenReceiver ($conversation_id: ID!, $message_id_2 : ID!){
#   UpdateLastSeenReceiver(conversation_id : $conversation_id , message_id_2 : $message_id_2){
#     conversation_id
#     sender
#     receiver
#     t_create
#     message_id_1
#     t_update_1
#     message_id_2
#     t_update_2
#   }
# }`

api_name = "UPDATE_LAST_SEEN_RECIEVER"                             #api import name
api_call_name = "UpdateLastSeenReceiver"                         #api call name


test_component_name = api_call_name+"MutationCCQT"                     #component name

input_list = """
conversation_id
message_id_2
"""


output_list = """
  conversation_id
    sender
    receiver
    t_create
    message_id_1
    t_update_1
    message_id_2
    t_update_2
"""

api_parameters = input_list.split()       #api input param
api_parameters_value = [elem + "_input" for elem in input_list.split()]    #api immput param value

api_returns = output_list.split()      #return values parameter by api
returns_list_value = [elem+"_output" for elem in output_list.split()] #return values by api
```

```python
def prop_list(pl):
    st = ""
    for elem in pl:
        st+= " "+elem+" ,"
    return st
    
def prop_type(pl):
    st = ""
    for elem in pl:
        st+= " "+elem+": string ,"
    return st
def create_responce_text(rlv,ar):
    st=""
    for elem in rlv:
        st += "{data."+ar+"."+elem + "}\n"+" "*10
    return st[:-11]


print(
 """export const """+test_component_name+""" = ({ """+prop_list(api_parameters)+""" }: { """+prop_type(api_parameters)+"""  }) => (
  <Mutation mutation={"""+api_name+"""} variables={{ """+prop_list(api_parameters)+""" }}>
    {(mutate, { loading, error, data }) => {
      if (loading) return <Text>Loading...</Text>
      if (error) return <Text>Error!</Text>
      if (data)
        return (
          <Text>
          """
    +
            create_responce_text(api_returns,api_call_name)
    +
          """
          </Text>
        )
      return (
        <View>
          <TouchableOpacity
            onPress={() =>
              mutate({
                variables: {
                  """+prop_list(api_parameters)+"""
                }
              })
            }
          />
        </View>
      )
    }}
  </Mutation>
)""")
```
